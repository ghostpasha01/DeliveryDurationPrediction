```python
from scipy import stats
import numpy as np

#finansal özelliklerin Z-puanlarını hesaplar ve bu Z-puanlarının 3'ten büyük olduğu yerleri (np.where(z>3)) döndürür.
z=np.abs(stats.zscore(data[financial_columns]))
print(np.where(z>3))
```

```python
# finansal özelliklerini içeren bir data_outliers veri seti oluşturur.
data_outliers=data[financial_columns][(z>3).any(axis=1)]
data.iloc[data_outliers.index]
```

aykırı örneklerin indekslerini (i[0]) alır ve bu indeksleri kullanarak orijinal data veri setindeki tüm özellikleri içeren
satırları (data.iloc[indices]) döndürür.

```python

outliers_row_indices=np.where(z>3)[0]
unique_values,counts=np.unique(outliers_row_indices,return_counts=True)
outliers_counts=np.array((unique_values,counts)).T

threshold=7

outliers=[x for x in outliers_counts if x[1]>threshold]
indices=[i[0] for i in outliers]
data.iloc[indices]
```

# Çok Değişkenli Aykırı Değer Tespiti

```python
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
```

inference değişkeninde, aykırı olmayan örnekler 1 olarak etiketlenirken, aykırı örnekler -1 olarak etiketlenir.
np.where(inference==-1) kullanarak aykırı örneklerin indekslerini (indices_if) belirler ve data.iloc[indices_if]
kullanarak orijinal data veri setindeki tüm özellikleri içeren satırları döndürür

```python
isolation_forest=IsolationForest()

isolation_forest=isolation_forest.fit(data[financial_columns])
inference=isolation_forest.predict(data[financial_columns])

indices_if= np.where(inference==-1)[0]
print("There are",len(indices_if),"members marked as outliers")
data.iloc[indices_if]
```

```python
# aynı şekilde aykırı değer tespit çalışması
local_outlier_factor=LocalOutlierFactor()
inference=local_outlier_factor.fit_predict(data[financial_columns])
indices_lof=np.where(inference==-1)[0]
print(" there are",len(indices_lof),"members marked as outliers")
data.iloc[indices_lof]
```

```python
common_indices=list(set(indices_if).intersection(set(indices_lof)))
data.iloc[common_indices]
```

```python
_=data.iloc[common_indices][financial_columns].hist(figsize=(19,13))
```
